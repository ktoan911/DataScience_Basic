```python
class Student:
    def __init__(self, studentID, name, grades):
        self.studentID = studentID
        self.name = name
        self.grades = grades

    def get_average(self):
        total = sum(self.grades)
        average = total / len(self.grades)
        return average

    def __str__(self):
        return f'Student ID: {self.studentID}, Name: {self.name}, Average Score: {self.get_average():.2f}'

def main():
    num = int(input('Enter the total number of students: '))
    students = []

    for _ in range(num):
        studentID = input("Enter student ID: ")
        name = input('Enter student name: ')
        grades = []

        for i in range(3):
            grade = float(input(f"Enter grade {i + 1} for {name}: "))
            grades.append(grade)

        s = Student(studentID, name, grades)
        students.append(s)

    students.sort(key=lambda student: student.get_average())

    print("\nStudents sorted by average grade in ascending order:")
    for student in students:
        print(student)

    remove_id = input("Enter student ID to remove: ")
    students = [student for student in students if student.studentID != remove_id]

    print("\nUpdated list of students after removing:")
    for student in students:
        print(student)

if __name__ == '__main__':
    main()
```

```python
class Person:
    def __init__(self, name, address, gender):
        self.name = name
        self.address = address
        self.gender = gender

class Employee(Person):
    def __init__(self, employee_id, name, address, gender, salary):
        super().__init__(name, address, gender)
        self.employee_id = employee_id
        self.salary = salary
        self.salary_history = [salary]

    def promote(self):
        # Increase salary by 5% and add to the salary history
        new_salary = self.salary * 1.05
        self.salary_history.append(new_salary)
        self.salary = new_salary

    def __str__(self):
        return f'Employee ID: {self.employee_id}, Name: {self.name}, Salary: ${self.salary:.2f}'


if __name__ == '__main__':
    num_employees = int(input('Enter the number of employees: '))
    employees = []

    for i in range(num_employees):
        employee_id = input(f'Enter employee ID for Employee {i + 1}: ')
        name = input('Enter name: ')
        address = input('Enter address: ')
        gender = input('Enter gender: ')
        salary = float(input('Enter salary in dollars: '))
        
        employee = Employee(employee_id, name, address, gender, salary)
        employees.append(employee)

    # Display the list of employees before promotion
    print("\nEmployees before promotion:")
    for employee in employees:
        print(employee)

    # Select two employees for a salary promotion
    for _ in range(2):
        promo_employee_id = input("Enter employee ID to promote: ")
        for employee in employees:
            if employee.employee_id == promo_employee_id:
                employee.promote()
                print(f"Promoted {employee.name} (Employee ID: {employee.employee_id})")

    # Display the updated list of employees after promotion
    print("\nEmployees after promotion:")
    for employee in employees:
        print(employee)
```
