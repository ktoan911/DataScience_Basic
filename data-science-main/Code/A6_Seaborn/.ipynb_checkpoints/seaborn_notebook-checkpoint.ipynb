### Distribution Plot
* Univariate distribution of observations (Phân phối đơn biến)

### KDE Plot
* A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analagous to a histogram. 
* KDE represents the data using a continuous probability density curve in one or more dimensions

```python
sns.kdeplot(data = tip_df.total_bill)
```

### Displot

* Tóm tắt phân phối thông tin
* Phân tích thống kê yêu cầu kiến ​​thức về phân bố các biến trong tập dữ liệu của bạn.
* Hàm seaborn `displot()` hỗ trợ một số cách tiếp cận để hiển thị sự phân bố. Chúng bao gồm các kỹ thuật cổ điển như biểu đồ và các phương pháp tính toán chuyên sâu như ước tính mật độ hạt nhân.

```python
sns.displot(data = tip_df, x = 'total_bill',col = 'time', kde = True)
# kde = true để phối hợp hipt và kde
# col phân chia bảng
```

### Box Plot
Box Plot is a good indication of how the values in the data are spread out ("minium", first quartile (Q1), median, third quartile (Q3), and “maximum”)
* Identify outliers
* Check if data is symmetrical
* Determine how tightly data is grouped

Resource: https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51 

-> Hiểu được giá trị phân bố thế nào

```python
sns.boxplot(data = tip_df, x = 'day', y = 'total_bill', hue = 'sex')
plt.legend(loc = 0) # loc là vị trí cụ thể của chú thích
```

### Facet Grid
* FacetGrid can be drawn with up to three dimensions: `row`, `col`, and `hue`.
* The main approach for visualizing data on this grid is with the FacetGrid.map() method. 

-> có thể vẽ được 3 chiều

```python
tip_df.head()
```

```python
tip_fg = sns.FacetGrid(data = tip_df, row = 'smoker', col = 'time') # Create a class instance của Facet Grid class
tip_fg.map(sns.scatterplot, 'total_bill', 'tip') # bên trong đồ thị ( chỉnh color, size các chấm)
```

```python
new_fg = sns.FacetGrid(data = tip_df, col = 'sex') # Create a class instance của Facet Grid class
new_fg.map(sns.scatterplot, 'total_bill', 'tip')
```

```python

new_fg = sns.FacetGrid(data = tip_df, col = 'sex',hue = 'smoker', col_order=['Female', 'Male'], palette = 'Set2', 
                       height = 4, aspect = 1.4) # Create a class instance class Facet Grid class

new_fg.map(sns.scatterplot, 'total_bill', 'tip', s = 100, edgecolor = 'b', alpha = .7) # bên trong đồ thị
#edgecolor : màu sắc của viền 
#s là size chấm tròn


#hàm map sẽ TƯƠNG ĐƯƠNG với 
kws = dict(s = 100, edgecolor = 'b', alpha = .7)
new_fg.map(sns.scatterplot, 'total_bill', 'tip', **kws) 

new_fg.add_legend()
```

## Project: Nutrious Data Visualisation on 80 cereal products

```python
#load dataset
cereal_df = pd.read_csv('./data/cereal.csv')
```

```python
cereal_df.head()
```

```python
cereal_df.isna().sum()
```

```python
# irrelevant fields
# loại bỏ những cột ko liên quan đến yêu cầu topic (ko liên quan đến giá trị dinh dưỡng)

field = ['shelf', 'weight', 'cups', 'rating']
cereal_df_new = cereal_df.drop(field, axis = 1) # drop cột nên axis = 1
```

```python
cereal_df_new.head()
```

```python
#corr
#chỉ nhận cột số

cereal_corr = cereal_df_new.select_dtypes(exclude=['object']).corr() # loại những cột có kiểu dữ liệu object
cereal_corr
```

```python
#nhận xét: đây là ma trận đối xứng có đường chéo chính full 1 -> chỉ cần lấy dữ liệu nửa tam giác dưới
# ones_like can build a maxtrix of boolean with the same shape of out data

ones_corr = np.ones_like(cereal_corr, dtype =bool)
ones_corr
```

```python
ones_corr.shape, cereal_corr.shape
```

```python
#np' s triu : return upper triangle matrix
mask = np.triu(ones_corr)
mask
```

```python
sns.heatmap(data = cereal_corr, mask = mask); # chỉ vẽ trên đoạn mask có giá trị bằng false
```

```python
#nhận thấy hàng calo và cột vitamin bị trống do mask cho đường chéo chính có giá trị true
```

```python
#nếu ko có mask
sns.heatmap(data = cereal_corr);
```
