## Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> quickly
* OO API -> advance

### Pyplot API 

```python
x = np.linspace(0,10, 1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'dashed',label = 'x') # vẽ đồ thị
plt.plot(x, np.cos(x), color = 'red', label = 'f(x)') # vẽ đồ thị

plt.title('F(x)') #đặt tên cho đồ thị

plt.xlabel('Bien x') # đặt label cho cột x
plt.ylabel('f(x))') # đặt lable cho cột y

plt.xlim([0,10]) # chỉ hiện x chạy từ 0 đến 10
plt.ylim([-1, 1]) # chỉ hiện y chạy từ -1 đến 1
# hoặc 
plt.axis([0, 10 , -1, 1]) #tương đương 2 dòng trên

plt.axis('equal') # vẽ sao cho trục x và y bằng nhau


plt.legend();  # Hiển thị chú thích

plt.show() # hiển thị đồ thị ( ko cần thiết cho jupyter notebook)
```

### OO API

```python
X =[1,2,3,4]
Y = [10, 8 ,6, 9]
```

```python
#set up plot
fig, ax = plt.subplots(figsize = (5,5))
ax.plot(X, Y);

ax.set(title = 'Simple plot', xlabel = 'x', ylabel = 'y',)
```

```python
ax.set(title = 'Simple plot', xlabel = 'x', ylabel = 'y')
```
