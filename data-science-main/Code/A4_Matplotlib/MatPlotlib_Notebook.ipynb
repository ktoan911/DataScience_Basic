## Pyplot API vs Object-Oriented (OO) API
* Pyplot API -> quickly
* OO API -> advance

### Pyplot API 

```python
x = np.linspace(0,10, 1000)
plt.plot(x, np.sin(x), color = 'blue', linestyle = 'dashed',label = 'x') # vẽ đồ thị
plt.plot(x, np.cos(x), color = 'red', label = 'f(x)') # vẽ đồ thị

plt.title('F(x)') #đặt tên cho đồ thị

plt.xlabel('Bien x') # đặt label cho cột x
plt.ylabel('f(x))') # đặt lable cho cột y

plt.xlim([0,10]) # chỉ hiện x chạy từ 0 đến 10
plt.ylim([-1, 1]) # chỉ hiện y chạy từ -1 đến 1
# hoặc 
plt.axis([0, 10 , -1, 1]) #tương đương 2 dòng trên

plt.axis('equal') # vẽ sao cho trục x và y bằng nhau


plt.legend();  # Hiển thị chú thích

plt.show() # hiển thị đồ thị ( ko cần thiết cho jupyter notebook)
```

```python
# OO API

#Prepare New Data 
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
```

```python
np.mean(x), np.mean(y)
```

```python
fig, ax = plt.subplots()
colors = rng.rand(100)
sizes = 100 * rng.rand(100)
plot = ax.scatter(x, y, s = sizes, c = colors, cmap = 'viridis', alpha =0.3)

# thêm color bar ở bên phải
fig.colorbar(plot)
```

# Example : Data Visualization - California Cities

```python
cities = pd.read_csv ('./data/california_cities.csv')
cities.head()
```

```python
# Extract lattitude and longtitude

lat = cities['latd']
long = cities['longd']
lat.head()
```

```python
type(lat)
```

```python
population, area = cities.population_total, cities.area_land_km2

#Plot using Pylot API
plt.figure(figsize = (8,6))
plt.scatter(long, lat, c = np.log10(population), cmap = 'viridis', s = area, alpha = 0.5, linewidth = 0) 
plt.axis('equal')
plt.xlabel('longlatitude')
plt.ylabel('latitude');
plt.colorbar(label = 'log$_{10}$(population)');

#Create a legend for cities'size 
area_rangeDefault = [50, 100, 300, 500]

for area in area_rangeDefault:
    plt.scatter([],[], s = area, c = 'k', alpha = 0.4,  label = str(area) + 'km$^2$')
plt.legend(labelspacing = 1, scatterpoints = 1, title = 'City Area')
```
